```python
import numpy as np
df['median_income_cat'] = pd.cut(df['median_income'],
                                 bins=[0.,2.,4.,6.,8.,np.inf],
                                 labels=[1,2,3,4,5])
df
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['median_income_cat']):
  strat_train = df.loc[train_index]
  strat_test = df.loc[test_index]
```

```python
x_train = strat_train.drop(columns=['median_house_value'])
y_train = strat_train['median_house_value']

x_test = strat_test.drop(columns=['median_house_value'])
y_test = strat_test['median_house_value']
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class feature_derivations(BaseEstimator, TransformerMixin):
  def __init__(self, add_bedrooms_per_room=True):
    self.add_bedrooms_per_room = add_bedrooms_per_room
  def fit(self, x, y=None):
    return self
  def transform(self, x):
    rooms_per_household = x[:,3]/x[:,6]
    population_per_household = x[:,5]/x[:,6]
    if self.add_bedrooms_per_room:
        bedrooms_per_room = x[:,4]/x[:,3]
        return np.c_[x, rooms_per_household, population_per_household, bedrooms_per_room]
    else:
        return np.c_[x, rooms_per_household, population_per_household]
```

```python
# just to check if the feature_derivations is working
attr_adder = feature_derivations(add_bedrooms_per_room=False)
new = attr_adder.transform(x_train.values)
new
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('featre_addition', feature_derivations(add_bedrooms_per_room=False)),
    ('scaler', StandardScaler())
])

# just to check if the pipeline is working
new = num_pipeline.fit_transform(x_train.drop(columns=['ocean_proximity']))
new
```

```python
num_variables = list(x_train.drop(columns=['ocean_proximity']))
cat_variables = ['ocean_proximity']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_variables),
    ('cat', OneHotEncoder(), cat_variables)
])

x_prep = full_pipeline.fit_transform(x_train)
x_prep
```

```python
from sklearn.linear_model import LinearRegression


lr = LinearRegression()
lr.fit(x_prep, y_train)
```
